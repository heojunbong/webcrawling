```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd 


# 웹 드라이버 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 현재 페이지의 댓글을 크롤링
channel_url = 'https://www.enuri.com/detail.jsp?modelno=101805210'

# implicitly_wait() 메서드로 기다리는 시간 설정

driver.implicitly_wait(3)

# 채널 페이지로 이동

driver.get(channel_url)
data=pd.DataFrame()
review_date=[]
score=[]
title=[]
review=[]
page_number = 1

# "다음 페이지" 버튼을 클릭하며 댓글 크롤링하는 루프

while True:
    p_elements_score = driver.find_elements(By.CLASS_NAME, "tx_aval")
    for p_element_score in p_elements_score:
        p_score_text = p_element_score.text
        score.append(p_score_text)
        print(p_score_text)

    page_number += 1
    button_elements = driver.find_elements(By.CSS_SELECTOR, "button.p_num")
    
    # 다중 조건문: page_number가 10으로 나누었을 때 나머지가 1이면 다음 버튼을 클릭
    
    if page_number != 1 and page_number % 10 == 1:
        next_button = driver.find_element(By.CLASS_NAME, "btn__next")
        next_button.click()
        time.sleep(1)
        button_elements = driver.find_elements(By.CSS_SELECTOR, "button.p_num")
    for button in button_elements:
        if int(button.text) == page_number:
            button.click()
            time.sleep(1)
            break
        
        
    else: # for~else문을 사용했습니다. 위 for문을 돌다가 break 명령어를 못만나게 되면, 이 else문으로 오게 됩니다. 반면 for문을 돌다가 break를 만났으면 이 else문으로 오지 않습니다.
        break # while True 탈출
```

```python
def remove_continuous(lst):
    result = []
    i = 0
    while i < len(lst): 
        if i < len(lst) - 7 and lst[i] == '5' and lst[i+1] == '5' and lst[i+2] == '5' and lst[i+3] == '4.8'and lst[i+4] == '4.9'and lst[i+5] == '4.7' and lst[i+6] == '5.0' and lst[i+7] == '4.8':
            i += 8
        else:
            result.append(lst[i])
            i += 1
    return result

b = remove_continuous(scores)
print(b)
```
